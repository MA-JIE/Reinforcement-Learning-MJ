# 二.Generative Adversarial Nets: GAN <p>
  基于博弈论,目的是找到达到纳什均衡的判别器网络和生成器网络 <p>
  生成器网络: 以一个潜在空间的随机向量作为输入,并将其解码为一张合成图像.就像伪造达芬奇画作的伪造者一样. <p>
  判别器网络: 以一张图像(真实的或合成的)作为输入,并预测该图像来自训练集还是来自生成器网络.就像鉴别赝品的鉴赏者一样.<p>
  

      GAN的优化过程不是求损失函数的最小值,而是保持生成与判别两股力量的动态平衡.

```python
import os
import torch
import torchvision
import torch.nn as nn
from torchvision import transforms
from torchvision.utils import save_image

#设备配置
torch.cuda.set_device(0)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

#定义超参数
latent_size = 64
hidden_size = 256
image_size = 784
num_epochs = 200
batch_size = 100
sample_dir = 'gan_samples'

if not os.path.exists(sample_dir):
    os.makedirs(sample_dir)

#image procesing
#Compose可以把一些转换函数组合在一起;Totensor()把一个取值范围是[0,255]的PIL.Image转换成tensor;Normalize是对张量进行归一化,
#表示全局平均值和方差,减均值,除以标准差,若有多个通道,则Normalize([m1,m2,m3],[n1,n2,n3])
trans = transforms.Compose([transforms.ToTensor(),transforms.Normalize([0.5],[0.5])])

#mnist data,初始时, download = True 用于下载数据,60000 = 100 x 600
mnist = torchvision.datasets.MNIST(root='data3',
                                  train = True,
                                  transform=trans,
                                  download= False)
#data loader
data_loader = torch.utils.data.DataLoader(dataset= mnist,
                                         batch_size= batch_size,
                                         shuffle=True)
```

```python
#构建判断器
#sigmoid(): 1/(1+exp(-x))
#LeakyReLU(): max(ax,x)
#ReLU(): max(0,x)
D = nn.Sequential(
    nn.Linear(image_size, hidden_size),
    nn.LeakyReLU(0.2),
    nn.Linear(hidden_size,hidden_size),
    nn.LeakyReLU(0.2),
    nn.Linear(hidden_size,1),
    nn.Sigmoid())
#构建生成器,相当于AVE中的解码器
#Tanh(): (1-exp(-2x) / (1+exp(2x)))
G = nn.Sequential(
    nn.Linear(latent_size, hidden_size),
    nn.ReLU(), 
    nn.Linear(hidden_size,hidden_size),
    nn.ReLU(),
    nn.Linear(hidden_size, image_size),
    nn.Tanh())

#把判别器和生成器迁移到GPU上
D = D.to(device)
G = G.to(device)

#定义判别器的损失函数交叉熵及优化器
criterion = nn.BCELoss()
d_optimizer = torch.optim.Adam(D.parameters(), lr = 0.0002)
g_optimizer = torch.optim.Adam(G.parameters(), lr = 0.0002)

#Clamp区间限定函数x限制在区间[min,max]内
def denorm(x):
    out = (x + 1) / 2
    return out.clamp(0,1)
#梯度置零
def reset_grad():
    d_optimizer.zero_grad()
    g_optimizer.zero_grad()
    
#开始训练
total_step = len(data_loader)
print(len(data_loader))
```

```python
for i, (images,_) in enumerate(data_loader):
    if(i<5):
        print(images.size())
#         print(len(iamges))
        images = images.reshape(100,-1)
        print(images[0])
    else:
        break
```

```python
for epoch in range(num_epochs):
    for i, (images, _) in enumerate(data_loader):
        images = images.reshape(batch_size,-1).to(device)
        #定义图像是真或假的标签
        real_labels = torch.ones(batch_size,1).to(device)
        fake_labels = torch.zeros(batch_size, 1).to(device)
        
        ########训练判别器######################
        
        #相当于告诉判别器真图像为真,假图像为假
        
        #定义判别器对真图片的损失函数
        outputs = D(images)
        d_loss_real = criterion(outputs,real_labels)
        real_score = outputs
        #定义判别器对假图片(由潜在空间点生成的图片)的损失函数,latent_size相当于image_size
        z = torch.randn(batch_size, latent_size).to(device)
        fake_images = G(z)
        outputs = D(fake_images)
        d_loss_fake = criterion(outputs, fake_labels)
        fake_score = outputs  
        
        #得到判别器总的损失函数,同时提高判别真和假的能力
        d_loss = d_loss_real + d_loss_fake
        #对生成器,判别器的梯度清零
        reset_grad()
        d_loss.backward()
        d_optimizer.step()
        
        
        ##########训练生成器#####################
        
        # 定义生成器对假图片的损失函数，这里我们要求
        #生成器生成的图片越来越像真图片，故损失函数中
        #的标签改为真图片的标签，即希望生成的假图片，
        #越来越靠近真图片
        z = torch.randn(batch_size, latent_size).to(device)
        fake_images = G(z)
        outputs = D(fake_images)
        
        g_loss = criterion(outputs, real_labels)
        
        #对生成器,判别器的梯度清零
        #反向传播以及运行生成器的优化器
        reset_grad()
        g_loss.backward()
        g_optimizer.step()
        if (i+1) % 200 == 0:
            print('Epoch [{}/{}], Step [{}/{}], d_loss:{:.4f}, g_loss: {:.4f}, D(x): {:.2f}, D(G(z)): {:.2f}'
                 .format(epoch, num_epochs, i+1, total_step, d_loss.item(), g_loss.item(),
                        real_score.mean().item(), fake_score.mean().item()))
    #保存真图片
    if(epoch + 1) == 1:
        images = images.reshape(images.size(0), 1,28,28)
        save_image(denorm(images[0]), os.path.join(sample_dir,'real_images.png'))
    #保存假图片
    fake_images = fake_images.reshape(images.size(0),1,28,28)
    save_image(denorm(fake_images[0]), os.path.join(sample_dir,'fake_images-{}.png'.format(epoch+1)))
    
#保存模型
torch.save(G.state_dict(),'G.ckpt')
torch.save(D.state_dict(), 'D.ckpt')
    
```
